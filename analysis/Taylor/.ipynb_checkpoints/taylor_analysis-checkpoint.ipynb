<h1>Table</h1>

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
plt.style.use('bmh')


df = pd.read_csv("../../data/raw/AAPL.csv")
df.tail()

```

<hr>

<h1>Visuals</h1>

<i>Dropping all columns other than 'Date' & 'Time' as those will be used to visualize the daily high</i>

```python
df.drop(columns=['Open', 'Low', 'Close', 'Adj Close', 'Volume'], inplace = True)
df.tail()
```

<b>A VISUALIZATION OF THE DAILY STOCK PRICE HIGH</b>

```python


plt.figure(figsize=(20,5))
plt.title('Daily High History')
plt.plot(df['High'])
plt.xlabel('Date', fontsize = 15)
plt.ylabel('Daily High', fontsize =15)
plt.show()
```

<h2>Bar plot</h2>

```python
import pandas as pd
import seaborn as sns
csvPath = "../../data/raw/AAPL.csv"
sns.set_theme(style="whitegrid")
tips = pd.read_csv(csvPath).sort_values("High", ascending = True).head(5)
ax = sns.barplot(x="Date", y="High", data=tips)
```

<h1>Scatterplot</h1>

```python
import pandas as pd
import seaborn as sns
csvPath = "../../data/raw/AAPL.csv"
sns.set_theme(style="whitegrid")
tips = pd.read_csv(csvPath).sort_values("High", ascending = True).head(5)
ax = sns.scatterplot(x="Date", y="High", data=tips)
```
