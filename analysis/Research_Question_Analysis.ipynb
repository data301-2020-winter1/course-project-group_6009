```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('bmh')


df = pd.read_csv("../data/raw/AAPL.csv")
```

```python
##Filtered the data from 2010-2019
```

```python
df['Date'] = pd.to_datetime(df['Date'])

start_date = '2010-01-01'
end_date = '2019-11-25'

mask = (df['Date'] > start_date) & (df['Date'] <= end_date)
df.dropna()
df = df.loc[mask]
df
```

```python
##Plotted the data for visual
```

```python
plt.figure(figsize=(18,6))
plt.title('Daily Close History')
plt.plot(df['Close'])
plt.xlabel('Trend', fontsize = 15)
plt.ylabel('Daily High', fontsize =15)
plt.show()
```

Explanation: The x-axis is the number number of days that the data has been inputted. From the graph, we can see that it is about ~9800 days of data. We took the most recent 2800 days = 7 years of data to display.
The y-axis is the daily high opening price.
From this data and graph, we can conclude that, without large external variables, 2020 will contain a record high for APPLE's data.

**We filtered the data from 1980 - 2019 to 2010 - 2019 so we can get a more accurate and better look at the trend of the closing stock. With this we will be able to make a more accurate prediction of the closing stock on a certain day in 2020**

Conclusion: From the graph, we can see that Apple's stock continues to rise. It is safe to assume that 2020 will continue to reach a new record high, assuming negligable external variables.
